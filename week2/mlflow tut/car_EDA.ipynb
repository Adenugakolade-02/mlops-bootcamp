```python
def plot_categorical_variable(data,categorical_features,target):
    for features in categorical_features:
        sns.boxplot(x=features,y=target,data=data)
        plt.show()
```

```python
plot_categorical_variable(data, categorical_variable,'price')
```

```python
def encode_categorical_variables(data,cat_variables,target):
    for features in cat_variables:
#     cat_map = {}
        encoded = data.groupby([features])[target].mean().sort_values().index
        cat_map = {k:i for i,k in enumerate(encoded)}
        data[features] = data[features].map(cat_map)
```

```python
encode_categorical_variables(data,categorical_variable,'price')
```

```python
def plot_categorical_variable(data,categorical_features,target):
    for features in categorical_features:
        sns.barplot(x=features,y=target,data=data)
        plt.show()
```

```python
### importing the evaluation metrics that will be used
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
```

```python
##creating evaluation metrics for our model

def eval_metrics(actual, pred):
        rmse = np.sqrt(mean_squared_error(actual, pred))
        mae = mean_absolute_error(actual, pred)
        r2 = r2_score(actual, pred)
        return rmse, mae, r2
```

```python
###creating a simple run

with mlflow.start_run():
    model = ElasticNet(alpha=0.01,l1_ratio=0.01, random_state=42)
    model.fit(train[features],y_train)
    y_pred = model.predict(test[features])
    rmse, mae, r2 = eval_metrics(y_test, y_pred)
    
    ##logging prameters manually
    
    mlflow.set_tag("model","ElasticNet")
    
    mlflow.log_param('alpha',0.01)
    mlflow.log_param('l1_ratio',0.01)
    mlflow.log_metric("rmse", rmse)
    mlflow.log_metric("r2", r2)
    mlflow.log_metric("mae", mae)
    mlflow.sklearn.log_model(model,"model")
```
